```python
X = train_keyword[['hashtags', 'keyword_target', 'cleaned_text']]
y = train_keyword['target']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

```python
# Parameter selection
ngram_range = (1,2)
min_df = 5
max_df = 0.5
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(ngram_range=ngram_range, min_df = min_df, max_df = max_df)
                        
X_train_tfidf = tfidf.fit_transform(X_train['cleaned_text'])
X_test_tfidf = tfidf.transform(X_test['cleaned_text'])
# labels_train = y
print(X_train_tfidf.shape)
print(X_test_tfidf.shape)
```

```python
X_train_tfidf_df = pd.DataFrame(X_train_tfidf.toarray(), columns=tfidf.get_feature_names())
X_test_tfidf_df = pd.DataFrame(X_test_tfidf.toarray(), columns=tfidf.get_feature_names())
```

```python
from sklearn.feature_extraction.text import CountVectorizer

countvec = CountVectorizer()
X_train_countvec = countvec.fit_transform(X_train['hashtags'])
X_test_countvec = countvec.transform(X_test['hashtags'])

print(X_train_countvec.shape)
print(X_test_countvec.shape)
```

```python
X_train_countvec_df = pd.DataFrame(X_train_countvec.toarray(), columns=countvec.get_feature_names())
X_test_countvec_df = pd.DataFrame(X_test_countvec.toarray(), columns=countvec.get_feature_names())
```

```python
X_train_all_features = X_train[['keyword_target']].reset_index(drop = True).merge(X_train_tfidf_df, left_index = True, right_index = True).merge(X_train_countvec_df, left_index = True, right_index = True)
X_test_all_features = X_test[['keyword_target']].reset_index(drop = True).merge(X_test_tfidf_df, left_index = True, right_index = True).merge(X_test_countvec_df, left_index = True, right_index = True)
```

```python
from sklearn.linear_model import LogisticRegression

logistic_reg = LogisticRegression()
logistic_reg.fit(X_train_all_features, y_train)
```

```python
from sklearn.metrics import accuracy_score

log_pred = logistic_reg.predict(X_test_all_features)
accuracy = accuracy_score(y_test, log_pred)*100
print(format(accuracy, '.2f'))
```

```python
from sklearn.metrics import accuracy_score

log_pred_train = logistic_reg.predict(X_train_all_features)
accuracy = accuracy_score(y_train, log_pred_train)*100
print(format(accuracy, '.2f'))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score


logistic_reg = LogisticRegression()
cv_score_logistic = cross_val_score(logistic_reg, X_train_all_features, y_train, scoring='accuracy', cv=5, n_jobs=-1)
cv_score_logistic.mean()
```
